```julia
using Plots, IJulia,DifferentialEquations, Interact,WebIO
function lorenz(du,u,p,t)
  α, β, γ= p  
 du[1] = α*(u[2]-u[1])
 du[2] = u[1]*(β-u[3]) - u[2]
 du[3] = u[1]*u[2] - (γ)*u[3]
end
u0 = [1.0;0.0;0.0]
tspan = (0.0,100.0)
  
α=10.0
β=28.0
γ=8.0/3.0 
p = [α, β, γ]
prob = ODEProblem(lorenz,u0,tspan,p)
sol = solve(prob)
plot(sol,vars=(1,2,3))
```
