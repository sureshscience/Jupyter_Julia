what sets Julia unique?

- The robust community has groups for using the language in computational biology, statistics, machine learning, image processing, differential equations, and physics, AI, ML and data sciences.<br><br>

### In all fields of science
- According to the 2019 survey, 73% of users and developers use Julia for research. 
- The most popular fields are statistics, engineering, machine learning, and computer science.

- The core language imposes Julia Base and the standard library is written in Julia 
- A rich language of types for constructing and describing objects, that can also optionally be used to make type declarations
- The ability to define function behavior across many combinations of argument types via multiple dispatch
- Automatic generation of efficient, specialized code for different argument types
- Good performance, approaching that of statically-compiled languages like C
